```python
import autogen
from autogen import ConversableAgent
from autogen import GroupChat
from autogen import GroupChatManager
```

```python
llm_config = {"model": "gpt-4o",
              "api_key": "Your_API_KEY"}
```

```python
proposal = '''
        We propose to make it mandatory for schools to give food to the students at school.
       '''
```

```python
head_advisor = ConversableAgent(
    name="head_advisor",
    system_message="You are the head advisor of a council. A proposal is given to you."
                    "You need to consult the advisors in your council to decide if the proposal should be absolutly rejected or accepted or could be accepted with modification."
                    "You need to also consult the advisors in your council and ask questions to make a comprehensive plan for execution."
                    " Gather all the insights from the council and provide a comperhensive excecution plan."
                    "Use TERMINATE when you want the chat rounds to finish.",
    is_termination_msg=lambda x: x.get("content", "").find("TERMINATE") >= 0,

    human_input_mode="NEVER",

    llm_config=llm_config,
)
```

```python
legal_advisor = ConversableAgent(
    name="legal_advisor",
    system_message="You are a legal advisor of a council. A proposal is given to you."
                    "You need to provide legal consultation to the head of council should the proposal be absolutly rejected or accepted or could be accepted with modification."
                    "You need to also provide legal advice to the head of the council to make modifications if nessasary and provide a comprehensive plan for execution." ,

    human_input_mode="NEVER",

    llm_config=llm_config,

)
```

```python
group_chat = GroupChat(
    agents=[legal_advisor, health_advisor, economic_advisor, logistics_and_operations_advisor, social_policy_advisor, education_advisor, environmental_advisor],
    #admin_name="head_advisor",
    messages=[],
    max_round=8,
    speaker_selection_method ="round_robin",

)
```

```python
group_chat_manager = GroupChatManager(
    groupchat=group_chat,
    llm_config={"config_list": [llm_config]},
)
```

```python
chat_result = head_advisor.initiate_chat(
    group_chat_manager,
    message=proposal,
    max_turns=2,
    summary_method="reflection_with_llm",
)
```
